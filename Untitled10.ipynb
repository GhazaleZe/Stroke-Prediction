```python
df1 = df.dropna()
df1['work_type'] = df1['work_type'].map({'Private':4, 'Self-employed':3, 'Govt_job':2, 'Never_worked':1})
df1['smoking_status'] = df1['smoking_status'].map({'formerly smoked':2, 'never smoked':1, 'smokes':3})
df1['gender'] = df1['gender'].map({'Male':2, 'Female':1})
df1['ever_married'] = df1['ever_married'].map({'Yes':1, 'No':0})
df1['Residence_type'] = df1['Residence_type'].map({'Urban':1, 'Rural':2})
df1
```

```python
from sklearn.model_selection import train_test_split
import random
#from sklearn.preprocessing import StandardScaler
#scaler = StandardScaler()
#print(scaler.fit(df1))
#x1 = scaler.transform(df1)
#df1 = pd.DataFrame(x1, columns = df1.columns)
df_train, df_test = train_test_split(df1, test_size = 0.2, random_state = 45)
ds_train = df_train.values
ds_test = df_test.values
X = np.concatenate((ds_train[:,1:9],ds_train[:,10:12]),axis = 1)
y = ds_train[:,9]
X_test = np.concatenate((ds_test[:,1:9],ds_test[:,10:12]),axis = 1)
y_test = ds_test[:,9]
X[np.isnan(X)] = np.median(X[~np.isnan(X)])
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X,y)
print('training result: ',reg.score(X,y)*100)
#y_pred = reg.predict(X_test)
print('testing result: ',reg.score(X_test,y_test)*100)
```

```python
y_pred = reg.predict(X_test)
temp = np.abs(y_pred - y_test)
temp.sum()/len(temp)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:\\term8\\datamining\\HW\\project\\healthcare-dataset-stroke-data.csv')
df = df[df.age > 18]
df
```

```python
df.isnull().sum()
```

```python
df1 = df
df1 = pd.get_dummies(df1, columns = ['work_type','smoking_status','gender','ever_married','Residence_type'])
df1 = df1.drop(columns = ['id'])
df1.age = (df1.age - df1.age.mean())/df1.age.std()
df1.avg_glucose_level = (df1.avg_glucose_level - df1.avg_glucose_level.mean())/df1.avg_glucose_level.std()
df_bmi_na = df1[df1.bmi.isnull()]
df1 = df1.dropna()
df1
```

```python
from sklearn.model_selection import train_test_split
import random
df_train, df_test = train_test_split(df1, test_size = 0.2, random_state = 45)
ds_train = df_train.values
ds_test = df_test.values
X = np.concatenate((ds_train[:,0:4],ds_train[:,5:20]),axis = 1)
y = ds_train[:,4]
X_test = np.concatenate((ds_test[:,0:4],ds_test[:,5:20]),axis = 1)
y_test = ds_test[:,4]
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X,y)
```

```python
y_pred = reg.predict(X)
temp = np.abs(y_pred - y)
temp.sum()/len(temp)
```
